```python
# What is a scatter plot ?
#ans- A scatter plot is a visual representation of how two variables relate to each other. You can use scatter plots to explore the relationship between two variables, for example by looking for any correlation between them.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))
plt.plot(x,y)
plt.title("Scatter Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
# Ques3- why is subplot()function is used ?
# ans - subplot() function in Python. subplot() function adds subplot to a current figure at the specified grid position. It is similar to the subplots() function however unlike subplots() it adds one subplot at a time. So to create multiple plots you will need several lines of code with the subplot() function.
import numpy as np 
x=np.array([0,1,2,3,4,5]) and y=np.array([0,100,200,300,400,500])
x=np.array([0,1,2,3,4,5]) and y=np.array([50,20,40,20,60,70])
x=np.array([0,1,2,3,4,5]) and y=np.array([10,20,30,40,50,60])
x=np.array([0,1,2,3,4,5]) and y=np.array([200,350,250,550,450,150])
fig = plt.figure()
ax = fig.add_subplot(projection ='2d')
plt.scatter(x,y)
plt.show()

```

```python
#Ques 4 - What is bar plot ? and why is it used ?
# ans -  Bar plots are a type of data visualization used to represent data in the form of rectangular bars. The height of each bar represents the value of a data point, and the width of each bar represents the category of the data. The Matplotlib library in Python is widely used to create bar plots.
import numpy as np 
company=np.array(["Apple","Microsoft","Google","AMD"])
profit=np.array([3000,8000,1000,10000])
plt.bar(company,profit)
plt.xlabel("Company name")
plt.ylabel("profit")
plt.title("bar graph ")
```

```python
import numpy as np 
company=np.array(["Apple","Microsoft","Google","AMD"])
profit=np.array([3000,8000,1000,10000])
plt.barh(company,profit)
plt.xlabel("Company name")
plt.ylabel("profit")
plt.title("Horizontal bar graph ")
```
